```python
# Assume df is your DataFrame with price data
# For example, df = pd.Series(your_price_data)

# Initialize results storage
profit_matrix = np.zeros((200, 200))

# Loop through all combinations of fast_ma and slow_ma
for fast in range(1, 201):
    for slow in range(1, 201):
        if fast < slow:
            fast_ma = vbt.MA.run(df, fast)
            slow_ma = vbt.MA.run(df, slow)

            # Generate entries and exits
            entries = fast_ma.ma_crossed_above(slow_ma)
            exits = fast_ma.ma_crossed_below(slow_ma)

            # Create portfolio
            pf = vbt.Portfolio.from_signals(df, entries, exits, init_cash=100)

            # Store total profit in the matrix
            profit_matrix[fast-1, slow-1] = pf.total_profit()

# Plot heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(profit_matrix, cmap="viridis", cbar_kws={"label": "Total Profit"})
plt.title("Heatmap of Total Profit for Fast MA and Slow MA")
plt.xlabel("Slow MA")
plt.ylabel("Fast MA")
plt.xticks(ticks=np.arange(0, 200, 20), labels=np.arange(1, 201, 20))
plt.yticks(ticks=np.arange(0, 200, 20), labels=np.arange(1, 201, 20))
plt.show()
```

```python
# Assume df is your DataFrame with price data
# For example, df = pd.Series(your_price_data)

results = []

# Loop through all combinations of fast_ma and slow_ma
for fast in range(1, 201):
    for slow in range(1, 201):
        # Ensure fast_ma is less than slow_ma to avoid redundant calculations
        if fast < slow:
            fast_ma = vbt.MA.run(df, fast)
            slow_ma = vbt.MA.run(df, slow)

            # Generate entries and exits
            entries = fast_ma.ma_crossed_above(slow_ma)
            exits = fast_ma.ma_crossed_below(slow_ma)

            # Create portfolio
            pf = vbt.Portfolio.from_signals(df, entries, exits, init_cash=100)

            # Append results
            results.append({
                "fast_ma": fast,
                "slow_ma": slow,
                "total_profit": pf.total_profit()
            })

# Convert results to a DataFrame for analysis
results_df = pd.DataFrame(results)

# Save or analyze the results
print(results_df.sort_values(by="total_profit", ascending=False))
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Prepare data for 3D plotting
fast_vals, slow_vals = np.meshgrid(np.arange(1, 201), np.arange(1, 201))
profit_matrix = np.where(fast_vals < slow_vals, profit_matrix, np.nan)

# Create a 3D surface plot
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(fast_vals, slow_vals, profit_matrix.T, cmap="viridis")

# Add labels and color bar
ax.set_title("3D Surface Plot of Total Profit for Fast MA and Slow MA")
ax.set_xlabel("Fast MA")
ax.set_ylabel("Slow MA")
ax.set_zlabel("Total Profit")
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=10, label="Total Profit")
plt.show()
```

```python
# Assume df is your DataFrame with price data
# Example: df = pd.Series(your_price_data)

# Initialize variables
best_sharpe_ratio = -np.inf
best_params = None
results = []

# Loop through all combinations of fast_ma and slow_ma
for fast in range(1, 201):
    for slow in range(1, 201):
        if fast < slow:
            fast_ma = vbt.MA.run(df, fast)
            slow_ma = vbt.MA.run(df, slow)

            # Generate entries and exits
            entries = fast_ma.ma_crossed_above(slow_ma)
            exits = fast_ma.ma_crossed_below(slow_ma)

            # Create portfolio
            pf = vbt.Portfolio.from_signals(df, entries, exits, init_cash=100)

            # Calculate Sharpe ratio
            sharpe_ratio = pf.sharpe_ratio()

            # Update best Sharpe ratio and parameters
            if sharpe_ratio > best_sharpe_ratio:
                best_sharpe_ratio = sharpe_ratio
                best_params = (fast, slow)

            # Store results for later analysis
            results.append({
                "fast_ma": fast,
                "slow_ma": slow,
                "sharpe_ratio": sharpe_ratio
            })

# Output the best Sharpe ratio and corresponding parameters
print(f"Best Sharpe Ratio: {best_sharpe_ratio}")
print(f"Best Parameters: Fast MA = {best_params[0]}, Slow MA = {best_params[1]}")

# Convert results to a DataFrame
results_df = pd.DataFrame(results)

# Analyze or save results
print(results_df.sort_values(by="sharpe_ratio", ascending=False).head())
```
